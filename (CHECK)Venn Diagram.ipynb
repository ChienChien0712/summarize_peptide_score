```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
plt.rcParams['figure.figsize'] = [10,10]
plt.rcParams['figure.dpi'] = 100
```

```python
file1 = '[condition 1] The protein-peptide table for Result from [neo4j] Start neo4j database'
file2 = '[condition 2] The protein-peptide table for Result from [neo4j] Start neo4j database'

f1 = pd.read_csv('./output2/%s/3.qcPass.csv'%file1)
f2 = pd.read_csv('./output2/%s/3.qcPass.csv'%file2)

print('[f1]listLength=%s, setLength=%s'%(len(f1['p.accession']),len(set(f1['p.accession']))))
print('[f2]listLength=%s, setLength=%s'%(len(f2['p.accession']),len(set(f2['p.accession']))))
```

```python
f1_T = set(f1[f1['qcPass']==True]['p.accession'])
f1_F = set(f1[f1['qcPass']==False]['p.accession'])

f2_T = set(f2[f2['qcPass']==True]['p.accession'])
f2_F = set(f2[f2['qcPass']==False]['p.accession'])

#(f1 ∩ f2)/f2
f1_Intersection_f2_T = len(f1_T&f2_T)

#(f2-f1)/f2
f2_Diff_f1 = len(f2_T - f1_T)

#(f1-f2)/f1
f1_Diff_f2 = len(f1_T - f2_T)

print('[Passed]')
print('pass in f1 = %s'%len(f1_T))
print('pass in f2 = %s'%len(f2_T))
print()
print('(f1 ∩ f2)/f2=%.2f%%(%s/%s)'%(f1_Intersection_f2_T*100/len(f2_T),f1_Intersection_f2_T,len(f2_T)))
print('(f2 - f1)/f2=%.2f%%(%s/%s)'%(f2_Diff_f1*100/len(f2_T),f2_Diff_f1,len(f2_T)))
print()
print('(f1 ∩ f2)/f1=%.2f%%(%s/%s)'%(f1_Intersection_f2_T*100/len(f1_T),f1_Intersection_f2_T,len(f1_T)))
print('(f1 - f2)/f1=%.2f%%(%s/%s)'%(f1_Diff_f2*100/len(f1_T),f1_Diff_f2,len(f1_T)))

venn2([f1_T, f2_T], set_labels = ('Condition 1                     ', '                 Condition 2'))
plt.title('QC Passed Protein',fontsize=20)
plt.show()
```
